### plots

```python
avgs = data.groupby('meter_id')[['CURRENT_TOTAL', 'VLN_AVG', 'FREQUENCY']].mean().reset_index()

import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
for col in ["VLN_AVG", "CURRENT_TOTAL", "FREQUENCY"]:
    fig = go.Figure()
    for mid, group in data.groupby("meter_id"):
        fig.add_trace(go.Scatter(
            x=group["timestamp"],
            y=group[col].round(4),
            mode="lines",
            name=f"Meter {mid}",
            hovertemplate=f"Meter {mid}<br>Time: %{{x}}<br>{col}: %{{y:.4f}}<extra></extra>"
        ))
    fig.update_layout(
        title=f"{col} over time",
        xaxis_title="Timestamp",
        yaxis_title=col,
        yaxis_tickformat=".4f",
        legend=dict(x=1.01, y=1),
        hovermode="closest",
        width=1200,
        height=500
    )
    fig.write_html(f"data/{col}.html")
    print(f"Saved data/{col}.html")
```
